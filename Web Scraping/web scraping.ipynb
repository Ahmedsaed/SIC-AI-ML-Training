```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
data = requests.get('http://goldpricez.com/eg/gram').text
soup = BeautifulSoup(data)
soup
```

```python
a = soup.find("div", attrs={"id":"price-list-sizes"}).find_all("td")
a
```

```python
weight = []
price = []
for i in range(len(a)):
    if i % 2 == 0:
        weight.append(a[i].text.strip())
    else:
        price.append(a[i].text.strip())
```

```python
data = pd.DataFrame({"Price": price, "Weight": weight})
data
```
