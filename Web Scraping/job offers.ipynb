# Imports & Configs

```python
import requests
from bs4 import BeautifulSoup
import warnings
from tabulate import tabulate
warnings.filterwarnings('ignore')
```

# Web Scraping

```python
def get_job_offers(job, page_number):
    print(f"searching for {job}, link: {'https://wuzzuf.net/search/jobs/?q=' + job.replace(' ', '+') + '&start=' + str(page_number)}")
    req = requests.get("https://wuzzuf.net/search/jobs/?q="+job.replace(' ', '+')).text
    soup = BeautifulSoup(req)
    job_offers = soup.find('div', attrs={'class': 'css-1gatmva e1v1l3u10'}).parent.find_all('div', attrs={'class': 'css-1gatmva e1v1l3u10'})
    return job_offers
```

```python
def get_details(offer):
    job_role = offer.find('a', attrs={"class": "css-o171kl"}).text
    job_link = "https://wuzzuf.net" + offer.find('a', attrs={"class": "css-o171kl"})['href']
    job_location = offer.find('span', attrs={"class": "css-5wys0k"}).text.strip()
    job_country = job_location.split()[-1]
    job_company = ' '.join(offer.find('a', attrs={"class": "css-17s97q8"}).text.split()[:-1])
    # job_description = BeautifulSoup(requests.get(job_link).text).find('h2')
    # print(job_description)

    return {'job_role': job_role,
            'job_link': job_link,
            'job_location': job_location,
            'job_country': job_country,
            'job_company': job_company,
            # 'job_description': job_description
            }
```

```python
jobs = ['Data analysis', 'machine learning', 'software testing']

for job in jobs:
    for i in range(3):
        job_offers = get_job_offers(job, i)
        for offer in job_offers:
            details = get_details(offer)
            print(tabulate(([key.replace('_', ' '), details[key]] for key in details.keys()), tablefmt="fancy_grid"))
```
