```python
def offer(row) :
    if row["gender"] == "Male" and row['SeniorCitizen'] == 1:
        return "Male _ se"
    elif row["gender"] == "Female" and row['SeniorCitizen'] == 1:
        return "Female _ se"
    elif row["gender"] == "Male" and row['SeniorCitizen'] == 0:
        return "Male _ jun"
    elif row["gender"] == "Female" and row['SeniorCitizen'] == 0:
        return "Female _ Jun"
    else:
        return "Other"
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df.columns
```

```python
gender_le = LabelEncoder()
Churn_le = LabelEncoder()
PaymentMethod_le = LabelEncoder()
```

```python
df['gender'] = gender_le.fit_transform(df["gender"])
```

```python
gender_le.classes_
```

```python
df
```

```python
df['Partner'] = Churn_le.fit_transform(df["Partner"])
df['Dependents'] = Churn_le.fit_transform(df["Dependents"])
df['Churn'] = Churn_le.fit_transform(df["Churn"])

df['PaymentMethod'] = PaymentMethod_le.fit_transform(df["PaymentMethod"])
```

```python
df.corr()
```

```python
df
```

```python
ten_min = df['tenure'].min()
ten_max = df['tenure'].max()
```

```python
df['tenure'] = (df['tenure'] - ten_min)/ (ten_max - ten_min)
```

```python
df
```

```python
tot_min = df['TotalCharges'].min()
tot_max = df['TotalCharges'].max()
```

```python
df['TotalCharges'] = (df['TotalCharges'] - tot_min)/ (tot_max - tot_min)
```

```python
tot_min1 = df['MonthlyCharges'].min()
tot_max1 = df['MonthlyCharges'].max()
df['MonthlyCharges'] = (df['MonthlyCharges'] - tot_min1)/ (tot_max1 - tot_min1)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop("Churn", axis = 1).values
y = df['Churn'].values 
```

```python
x_train,x_test ,y_train,y_test = train_test_split(x,y , test_size= 0.25 , random_state=42)
```
